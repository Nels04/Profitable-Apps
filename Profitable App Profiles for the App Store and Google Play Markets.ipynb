### Removing Duplicate entries

If you explore the Google Play data set long enough or look at the discussions section, you'll notice some apps have duplicate entries. Below we notice that there are 1,181 cases where an app occurs more than once in the Google store and 0 in the Apple store.

```python
# checking duplicates for both data set
def duplicate(dataset): # to take any dataset
    duplicate_apps = [] # for duplicates apps
    unique_apps = [] # for unique apps
    for app in dataset:
        name = app[0] # store the name of the app
        if name in unique_apps:
            duplicate_apps.append(name)
        else:
            unique_apps.append(name)
    return duplicate_apps
    return unique_apps
            
android_duplicate = duplicate(android_file)
ios_duplicate = duplicate(ios_file)

print('Number of duplicate adnroid apps:', len(android_duplicate))
print('\n')
print('Examples of duplicate android apps:', android_duplicate[:5])
print('\n')
# checking ios apps
print('Number of duplicate ios apps:', len(ios_duplicate))
```

For instance, Instagram has four duplicate entries:

```python
# Printing Instagram for Android
for app in android_file:
    name = app[0]
    if name == 'Instagram':
        print(app)
        print('\n')
```

If you examine the rows we printed for the Instagram app above, the main difference happens on the fourth position of each row, which corresponds to the number of reviews (under the column titled 'Reviews'. The different numbers show the data was collected at different times.

We can use this information to build a criterion for removing the duplicates. The higher the number of reviews, the more recent the data should be. Rather than removing duplicates randomly, we'll only keep the row with the highest number of reviews and remove the other entries for any given app, because the higher the number of reviews, the more reliable the ratings.


```python
# create a dictionary where each key is the unique app name
# and the value is highest number of reviews for that app
reviews_max = {}

for app in android_file:
    name = app[0]
    n_reviews = float(app[3])
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
        

```

Remember previously we found out that there are 1181 duplicate entries in the Google store data set. So the length of our unique apps should be the difference between our dataset and 1181

```python
print('Expected length is:', len(android_file) - 1181)
print('Actual length is:', len(reviews_max))
```

```python
android_clean = [] # our new cleaned data set. Unique apps
already_added = [] # app names

for app in android_file:
    name = app[0]
    n_reviews = float(app[3])
    if (reviews_max[name] == n_reviews) and (name not in already_added):
        android_clean.append(app)
        already_added.append(name)
        
```

Now let's see and confirm that our new and cleaned data set has 9659 rows.

```python
print(explore_data(android_clean, 0, 3, True))
```

## Removing non-English Apps

#### Part 1

Remember that our company only builds apps for English speaking audience. That's why we have to make sure that our data sets don't include any non-English app. Below are examples of non-English apps from both data sets.

```python
print('From iOS:', ios_file[813][1])
print('\n')
print('From Android:', android_clean[4412][0])
```

We'll need to remove apps like that and one way to go about it is to remove each app with a name containing a symbo that is not commonyl used in the English text.

English text usually includes letters from the English alphabet, numbers composed of digits from 0 to 9, punctuation marks (., !, ?, ;), and other symbols (+, *, /).

The numbers corresponding to the characters we commonly use in an English text are all in the range 0 to 127, according to the ASCII (American Standard Code for Information Interchange) system. Based on this number range, we can build a function that detects whether a character belongs to the set of common English characters or not. If the number is equal to or less than 127, then the character belongs to the set of common English characters. If an app name contains a character that is greater than 127, then it probably means that the app has a non-English name.

```python
# A function to check if a character ASCII number is greater than
# 127 or not. 
def is_english(a_string):
    for character in a_string:
        if ord(character) > 127:
            return False
        
    return True
    
print(is_english('Instagram'))
print(is_english('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))
print(is_english('Docs To Go‚Ñ¢ Free Office Suite'))
print(is_english('Instachat üòú'))
```

The function seems to work fine, but some English app names use emojis or other symbols (‚Ñ¢, ‚Äî (em dash), ‚Äì (en dash), etc.) that fall outside of the ASCII range. Because of this, we'll remove useful apps if we use the function in its current form.

For example: The TM symbols and emoji falls way above 127, that is why our is_english() function returned false for 'Docs To Go‚Ñ¢ Free Office Suite' and 'Instachat üòú' even though they're English apps.

```python
print(ord('‚Ñ¢'))
print(ord('üòú'))
```

If we're going to use the function we've created, we'll lose useful data since many English apps will be incorrectly labeled as non-English. To minimize the impact of data loss, we'll only remove an app if its name has more than three characters with corresponding numbers falling outside the ASCII range.

```python
# Modifying our is_english function
def is_english(a_string):
    non_ascii = 0
    for character in a_string:
        if ord(character) > 127:
            non_ascii += 1 # will increment by 1 everytime
            # there's a non_english character (outside of 127)
            
    if non_ascii > 3: # if there are more than 3 non_english character
        return False
    else:
        return True
    
print(is_english('Docs To Go‚Ñ¢ Free Office Suite'))
print(is_english('Instachat üòú'))
print(is_english('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))
```

Now we can use the modified is_english() function to filter out non-English apps

```python
android_english = [] # to store English android apps
ios_english = [] # to store English Apple apps

# Check for Android
for app in android_clean:
    name = app[0]
    if is_english(name):
        android_english.append(app)
        
# Check for ios
for app in ios_file:
    name = app[1]
    if is_english(name):
        ios_english.append(app)
        
print(explore_data(android_english, 0, 2, True))
print('\n')
print(explore_data(ios_english, 0, 2, True))
```

Now, below we'll build 2 functions we can use to analyze the frequency tables:
   * One function to generate frequency tables that show percentages
   * Another function we can use to display the percentages in a descending order

```python
# Function to generate frequency table in %
def freq_table(dataset, index):
    table = {}
    total = 0
    
    for row in dataset:
        total += 1
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1
            
    table_percentages = {}
    for key in table:
        percentage = (table[key] / total) * 100
        table_percentages[key] = percentage
    return table_percentages


# Function to display the % in a descending order       
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
        
```

### Most popular apps by Genre in the App Store

### Most Popular apps by Genre on Google Play

We have the data about the number of app installs for Google Play. However the install numbers are not precise enough. Down below we can see that the values are open-ended: 100+, 1000+, etc

```python
display_table(android_final, 5)
```

For instance, we don't know whether an app with 100,000+ installs has 100,000 installs, 200,000, or 350,000. However, we don't need very precise data for our purposes ‚Äî we only want to get an idea which app genres attract the most users, and we don't need perfect precision with respect to the number of users.

We're going to leave the numbers as they are, which means that we'll consider that an app with 100,000+ installs has 100,000 installs, and an app with 1,000,000+ installs has 1,000,000 installs, and so on.

To perform computations, however, we'll need to convert each install number to float ‚Äî this means that we need to remove the commas and the plus characters, otherwise the conversion will fail and raise an error. We'll do this directly in the loop below, where we also compute the average number of installs for each genre (category).

```python
android_category = freq_table(android_final, 1)

for category in android_category:
    total = 0
    len_category = 0
    for app in android_final:
        category_app = app[1]
        if category_app == category:
            n_installs = app[5]
            n_installs = n_installs.replace('+', '')
            n_installs = n_installs.replace(',', '')
            total += float(n_installs)
            len_category += 1
    avg_installs = total / len_category
    print(category, ':', avg_installs)
```

Communication apps seems to have to most installs. Keep in mind that there apps in that category with billion of users: WhatsApp, Facebook Messenger, Gmail, etc). The number is heavily skewed by these apps. 

```python
# Google play communication apps
for app in android_final:
    if app[1] == 'COMMUNICATION' and (app[5] == '1,000,000,000+' 
                                     or app[5] == '500,000,000+'
                                     or app[5] == '100,000,000+'):
        print(app[0], ':', app[5])
```

The numbers will significantly go down if we remove the apps above. 

```python
# removing 100M+ install apps
under_100m = []
for app in android_final:
    n_installs = app[5]
    n_installs = n_installs.replace('+', '')
    n_installs = n_installs.replace(',', '')
    if app[1] == 'COMMUNICATION' and (float(n_installs) < 100000000):
        under_100m.append(float(n_installs))
     
print('Average install of apps under 100M:', sum(under_100m) / len(under_100m)) 
```
